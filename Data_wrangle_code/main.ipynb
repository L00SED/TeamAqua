```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('UCMR3_reduced.csv')
```

```python
df.head(10)
```

```python
df['Contaminant'].unique()
```

```python
#output dataset for number of population impacted by each contaminant
df_detect = df.loc[df['indices']>0]
import numpy as np
#df_detect.drop_duplicates(subset=['PWSID','Contaminant','POPULATION_SERVED_COUNT']).shape()
df_detect_unique = df_detect.reindex(columns=['PWSID','Contaminant','POPULATION_SERVED_COUNT']).drop_duplicates()
df_detect_unique.reindex(columns=['Contaminant','POPULATION_SERVED_COUNT']).groupby(['Contaminant']).sum().to_csv('pop_detect.csv')
```

```python
subset = df.reindex(columns=['PWSID','Contaminant','AnalyticalResultValue','indices'])
```

```python
#aggregate by pwsid
f={'AnalyticalResultValue':['max','count'],'indices':['max','count']}
ucmr_by_pwsid = subset.groupby(["PWSID",'Contaminant']).agg(f).unstack()
```

```python
ucmr_by_pwsid.head(10)
```

```python
#output by pws result, manually remove empty cells from stack/unstack process, read in ucmrByPwsid_e.csv
ucmr_by_pwsid.to_csv('ucmrByPwsid.csv')
```

```python
#aggregate by zip
zip = pd.read_table('UCMR3_Zipcodes.txt')
ucmr_by_pwsid = pd.read_csv('ucmrByPwsid_e.csv')
```

```python
zip['PWSID'] = [ i.lstrip('0') for i in zip['PWSID'] ]
```

```python
result = pd.merge(ucmr_by_pwsid, zip, on='PWSID')
```

```python
#aggregate zip to county
#data source: US department of housing and urban development 
#https://www.huduser.gov/portal/datasets/usps_crosswalk.html
county = pd.read_csv('ZIP_COUNTY_032017.csv')
county['ZIPCODE']=[str(i).zfill(5) for i in county['ZIP']]
subset = county.reindex(columns=['ZIPCODE','COUNTY'])
subset.to_csv('zipcode_county_map.csv')
```

```python
byzip1=pd.concat([result.filter(regex=("count.*")), result['ZIPCODE']], axis=1).groupby(["ZIPCODE"]).sum()
```

```python
byzip2=pd.concat([result.filter(regex=("max.*")), result['ZIPCODE']], axis=1).groupby(["ZIPCODE"]).max()
```

```python
pd.concat([byzip1,byzip2],axis=1).to_csv('ucmrByZip.csv')
```

```python
ucmr_by_zip = pd.read_csv('ucmrByZip.csv')
```

```python
ucmr_by_zip['ZIPCODE'] = [ i.rstrip('-') for i in ucmr_by_zip['ZIPCODE'] ]
```

```python
ucmr_by_zip['ZIPCODE'] = [str(i).zfill(5) for i in ucmr_by_zip['ZIPCODE']]
```

```python
result2 = pd.merge(ucmr_by_zip, subset, on='ZIPCODE')
```

```python
result2.head(10)
```

```python
#result2.groupby(["COUNTY"]).max().to_csv('ucmrByCounty.csv')

byct1=pd.concat([result2.filter(regex=("count.*")), result2['COUNTY']], axis=1).groupby(["COUNTY"]).sum()
byct2=pd.concat([result2.filter(regex=("max.*")), result2['COUNTY']], axis=1).groupby(["COUNTY"]).max()
pd.concat([byct1,byct2],axis=1).to_csv('ucmrByCounty.csv')
```

```python
ucmr_by_county = pd.read_csv('ucmrByCounty.csv')
```

```python
#add county name to the dataset
#data source: the national bureau of economic research
#http://www.nber.org/data/ssa-fips-state-county-crosswalk.html
countyname = pd.read_csv('ssa_fips_state_county2017.csv').reindex(columns=['fipscounty','county','state'])
```

```python
countyname.columns = ['COUNTY', 'countyname','state']
ucmr_by_county2 = pd.merge(ucmr_by_county, countyname, on='COUNTY')
```

```python
ucmr_by_county2.countyname = ucmr_by_county2.countyname.str.title()
```

```python
ucmr_by_county2.head(10)
```

```python
#row sum
import numpy as np

ucmr_by_county2['indices.sum'] = np.nansum(ucmr_by_county2.filter(regex=("indices.max*")), axis=1)
ucmr_by_county2['counts.sum'] = np.nansum(ucmr_by_county2.filter(regex=("indices.count*")), axis=1)
```

```python
ucmr_by_county2['voc.indices.avg'] = ucmr_by_county2[['indices.max.1,2,3-trichloropropane',
                                        'indices.max.1,3-butadiene',
                                        'indices.max.chloromethane',
                                        'indices.max.1,1-dichloroethane',
                                        'indices.max.bromomethane',
                                        'indices.max.HCFC-22',
                                        'indices.max.Halon 1011']].apply(lambda row: np.nanmean(row),axis=1)
ucmr_by_county2['soc.indices.avg'] = ucmr_by_county2[['indices.max.1,4-dioxane']].apply(lambda row: np.nanmean(row),axis=1)
ucmr_by_county2['metal.indices.avg'] = ucmr_by_county2[['indices.max.vanadium',
                                                       'indices.max.molybdenum',
                                                       'indices.max.cobalt',
                                                       'indices.max.strontium',
                                                       'indices.max.chromium',
                                                       'indices.max.chromium-6']].apply(lambda row: np.nanmean(row),axis=1)
ucmr_by_county2['dbp.indices.avg'] = ucmr_by_county2[['indices.max.chlorate']].apply(lambda row: np.nanmean(row),axis=1)
ucmr_by_county2['pfas.indices.avg'] = ucmr_by_county2[['indices.max.PFOS',
                                                       'indices.max.PFOA',
                                                       'indices.max.PFNA',
                                                       'indices.max.PFHpA',
                                                       'indices.max.PFHxS',
                                                       'indices.max.PFBS']].apply(lambda row: np.nanmean(row),axis=1)
ucmr_by_county2['horm.indices.avg'] = ucmr_by_county2[['indices.max.17-beta-estradiol',
                                                       'indices.max.17-alpha-ethynylestrad',
                                                       'indices.max.estriol',
                                                       'indices.max.estrone',
                                                       'indices.max.equilin',
                                                       'indices.max.testosterone',
                                                       'indices.max.4-androstene-3,17-dion']].apply(lambda row: np.nanmean(row),axis=1)
```
